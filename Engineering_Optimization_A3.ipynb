```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# === Q1 ===
x = np.linspace(0, 8, 400)
rhos = [0.5, 1, 3]
plt.figure(figsize=(6,4))
for rho in rhos:
    f = np.where(x < 5, x + rho*(5-x)**2, x)
    plt.plot(x, f, label=f"ρ = {rho}")
plt.axvline(5, color='k', linestyle='--', linewidth=0.8)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Q1: Penalized function for different ρ values")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig("Q1_plot.png", dpi=1000)
plt.show()
plt.close()
```

```python
# === Q2 ===
x = np.linspace(-10, 15, 400)
rhos = [0.1, 1, 5]
plt.figure(figsize=(6,4))
for rho in rhos:
    F = (x - 5)**3 + rho*np.maximum(0, 10 - x)**2
    plt.plot(x, F, label=f"ρ = {rho}")
plt.axvline(10, color='k', linestyle='--', linewidth=0.8)
plt.xlabel("x")
plt.ylabel("F(x)")
plt.title("Q2: Quadratic penalty failure for cubic objective")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig("Q2_plot.png", dpi=300)
plt.show()
plt.close()
```

```python
# === Q3 (a) ===
mu = np.linspace(0, 5, 200)
for h in [0.5, 1, 2]:
    L = 1 + mu*h
    plt.plot(mu, L, label=f"h(x)={h}")
plt.xlabel("μ")
plt.ylabel("L(μ, x)")
plt.title("Q3(a): L(μ, x) for h(x) > 0")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig("Q3_plot1.png", dpi=300)
plt.show()
plt.close()
```

```python
# === Q3 (c) ===
mu = np.linspace(-2, 4, 200)
h = 1
L_valid = 1 + mu*h
plt.figure(figsize=(6,4))
plt.plot(mu, L_valid, label="L(μ, x) with h(x)>0")
plt.axvspan(-2, 0, color='red', alpha=0.1, label="Invalid (μ < 0)")
plt.axvspan(0, 4, color='green', alpha=0.1, label="Valid (μ ≥ 0)")
plt.xlabel("μ")
plt.ylabel("L(μ, x)")
plt.title("Q3(c): Effect of μ sign on constraint satisfaction")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig("Q3_plot2.png", dpi=300)
plt.show()
plt.close()
```
